```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def add_annotation(plot, feature, orient,bar_pct, bar_value):
    total = feature.count()
    
    if orient == 'v':
        for p in plot.patches:
            x = p.get_x() + p.get_width() / 2 - 0.1
            y = p.get_y() + p.get_height() / 2   
            
            if bar_pct:
                percentage = '{:.1f}%'.format(100 * p.get_height()/total)
                plot.annotate(percentage, (x, y), size = 12)
            if bar_value:
                value = p.get_height()
                plot.annotate(value, (x, y), size = 12)
    else:
        for p in plot.patches:
            x = p.get_x() + p.get_width() / 2 - 20
            y = p.get_y() + p.get_height() / 2
            
            if bar_pct:
                percentage = '{:.1f}%'.format(100 * p.get_width()/total)
                plot.annotate(percentage, (x, y), size = 12)
            if bar_value:
                value = p.get_width()
                plot.annotate(value, (x, y), size = 12)
                
def my_countplot(data, x=None, y=None,  bar_pct=False, bar_value=False):
    ax = sns.countplot(x=x, y=y, data=data)
    if bar_pct or bar_value:
        orient = 'h' if x is None else 'v'
        add_annotation(ax, titanic['class'], orient, bar_pct, bar_value)
```

## Proposed Countplot Behaviour

### Showing the bar percentage

We can show the bar percentages.

```python
my_countplot(titanic, x="class", bar_pct=True)
```

```python
my_countplot(titanic, y="class", bar_pct=True)
```

### Showing the Bar Value

We can show the bar values

```python
my_countplot(titanic, x="class", bar_value=True)
```

```python
my_countplot(titanic, y="class", bar_value=True)
```
